```python

""" 
Stack on Thu May 19 13:11:12 2020

"""

class stack():
    
    def __init__(self, size):
        self.value = []
        self.top = -1
        self.size = size
    
    def push(self, element):
        if not self.isFull():
            self.top += 1
            self.value.append(element)
        else:
            return "Stack Overflow"
    

    def pop(self):
        if self.isEmpty():
            return "Stack Underflow"
        else:
            temp = self.value[self.top]
            del[self.value[self.top]]
            self.top -= 1
            return temp
    
    def isEmpty(self):
        if (self.top < 0):
            return True
        return False
    
    def isFull(self):
        if(self.top == (self.size - 1)):
            return True
        return False
    
    def getSize(self):
        return self.size
    
    def getTop(self):
        return self.top
    
    def peek(self, position):
        if not isEmpty():
            if (position-1) <= self.getTop():
                return self.value[position-1]
            return "out of bounds"
        else:
            return "out of bounds"
        
    def show(self):
        return tuple(self.value)
    
    
# mystack = stack(10)

# print(mystack.isEmpty())
# print(mystack.isFull())
# print(mystack.getTop())
# print(mystack.getSize())
# mystack.push(90)
# mystack.push(91)
# mystack.push(92)
# mystack.push(93)
# mystack.push(94)
# print(mystack.isEmpty())
# print(mystack.isFull())
# print(mystack.getTop())
# print(mystack.getSize())
# print(mystack.peek(5))
# mystack.show()
```
