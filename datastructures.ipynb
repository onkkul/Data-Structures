```python

""" 
Queue on Thu May 20, 9:46:12 2020

"""

#===========================================================================
# Queue using two stacks
# Problem is once you begin enque, you cannot deque untill all insertions
# are complete and vise versa
#===========================================================================

# size = int(input("enter size"))
# stack_1 = stack(size)
# stack_2 = stack(size)

# for x in range(size):
#     stack_1.push(int(input("enter element: ")))

# for x in range(size):
#     stack_2.push(stack_1.pop())

# # stack_2 is now a queue
# for x in range(size):
#     print(stack_2.pop())

#===========================================================================




#===========================================================================
# Que that solves above issue
# using a circular que
#===========================================================================

class CircularQueue:
    
    def __init__(self, size):
        self.front = 0
        self.values = [None]*size
        self.rear = 0
        self.Maxsize = size
        self.Currsize = 0
        return

    def isEmpty(self):
        if self.Currsize == 0:
            return True
        return False
    
    def isFull(self):
        if self.Currsize == self.Maxsize:
            return True
        return False
    
    def getPointers(self):
        return {"Front":self.front, "Rear": self.rear, 
                "size": self.Currsize, "Maxsize": self.Maxsize}
    
    def show(self):
        if self.front < self.rear:
            return self.values[self.front:self.rear:]
        else:
            temp = self.values[self.front::]
            temp.extend(self.values[:self.rear:])
            return temp #self.values

    def enque(self, element):
        if self.isFull():
            return "Queue Full"
        else:
            self.Currsize +=1
            self.values[self.rear] = element
            self.rear += 1
            self.rear = self.rear%self.Maxsize
        return
            
    def deque(self):
        if self.isEmpty():
            return "Queue Empty"
        else:
            self.Currsize -= 1
            temp = self.values[self.front]
            self.values[self.front] = None
            self.front += 1
            self.front = self.front%self.Maxsize
        return temp

    def peek(self, position):
        if self.isEmpty():
            return "Queue Empty"
        if position > self.Maxsize or position < 0:
            return "Index out of bound"
        position = self.front + position - 1
        position = position % self.Maxsize
        return self.values[position]



#=================================Driver Code=====================================
# myque = CircularQueue(6)
# print(myque.isEmpty())
# print(myque.isFull())
# myque.enque(90)
# myque.enque(91)
# myque.enque(92)
# myque.enque(93)
# myque.enque(94)
# myque.enque(95)
# print(myque.enque(96))
# print(myque.show())
# print(myque.deque(),myque.getPointers())
# print(myque.deque(),myque.getPointers())
# print(myque.deque(),myque.getPointers())
# print(myque.deque(),myque.getPointers())
# print(myque.show())
# myque.enque(23)
# myque.enque(24)
# myque.enque(25)
# myque.deque()
# print(myque.show())
# print(myque.isEmpty())
# print(myque.isFull())

# print(myque.peek(5))
#=================================================================================
```

```python

""" 
LinkedLists on Thu May 21 15:13:12 2020

"""

class Node():
    
    def __init__(self, value):
        self.previous = None
        self.value = value
        self.next = None
    
    def __repr__(self):
        return f"previous: {self.previous.value}, current: {self.value}, next: {self.next.value}"

class LinkedList():
    
    def __init__(self):
        self.head = None
        self.tail = None
    
    def getMiddle(self):
        mid = None
        temp_head = self.tail
        while temp_head.next != self.head:
            mid = temp_head
            temp_head = temp_head.next.next
        return mid
    
    def insertMiddle(self, value):
        newNode = Node(value)
        mid = self.getMiddle()
        newNode.next = mid.next
        mid.next = newNode
        newNode.previous = mid

    
    def insert(self, value, _from = "head"):
        newNode = Node(value)
        if self.head == None:
            self.head = newNode
            self.tail = newNode
        else:
            self.head.next = newNode
            self.tail.previous = newNode
            newNode.previous = self.head
            newNode.next = self.tail

        if _from == "head":
            self.head = newNode
        elif _from == "tail":
            self.tail = newNode
        
   
    def traverse(self, _from = "tail"):
        values = []
        if _from == "tail":
            start = self.tail
            while start != self.head:
                values.append(start.value)
                start = start.next
            values.append(start.value)
        elif _from == "head":
            start = self.head
            while start != self.tail:
                values.append(start.value)
                start = start.previous
            values.append(start.value)
        return values

    
    def delete(self, count = 1, _from = "tail"):
        
        if _from == "mid":
            start = self.getMiddle()
            end = start
            for x in range(count):
                end = end.next
            start.previous.next = end
            end.previous = start.previous
            return
        
        for x in range(count):
            if _from == "head":
                self.head = self.head.previous
            if _from == "tail":
                self.tail = self.tail.next
            self.head.next = self.tail
            self.tail.previous = self.head

    def seek(self, position, _from = "tail"):
        values = self.traverse(_from)
        return values[position-1]

#=================================Driver Code=====================================
# myList = LinkedList()
# myList.insert(11)
# myList.insert(12)
# myList.insert(13)
# myList.insert(14)
# myList.insert(15)
# myList.insert(16)
# myList.insertMiddle(99)

# print(myList.traverse(_from = "tail"))
# myList.delete(count = 3, _from = "mid")
# myList.traverse(_from = "tail")
# print(myList.seek(5))
#=================================================================================
```
